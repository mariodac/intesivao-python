# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Abrindo o navegador e baixando a base de dados

```python
import pyautogui
import time
import os

pyautogui.PAUSE = 2

pyautogui.press("win") # abre menu windows
pyautogui.write("chrome") #pesquisa por chrome
pyautogui.press("enter") # aperta enter para abrir o chrome
time.sleep(4)
pyautogui.hotkey("win", "up") # para maximizar janela
time.sleep(2)
pyautogui.hotkey("ctrl", "l")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema") # digite o link do sistema
pyautogui.press("enter")  # aperta enter para entrar
time.sleep(4)
pyautogui.click(x=-771, y=343) # clica no campo de login
pyautogui.write("login") # digita o login
pyautogui.press("tab") # passar para o campo de senha
pyautogui.write("password") #digita a senha
pyautogui.click(x=-668, y=493) # clica no botão para entrar
time.sleep(4)
pyautogui.click(x=-1044, y=311) # clica no arquivo
pyautogui.click(x=-862, y=164) # clica no botão de baixar

pasta_downloads = os.path.expanduser("~") + "/Downloads" # obtem o diretorio de download
while True:
    time.sleep(1) 
    # verifica se ainda existe download em processo
    if any(fname.endswith('.crdownload') for fname in os.listdir(pasta_downloads)):
        print('O download ainda está em andamento')
    else:
        print('O download foi concluído')
        break
arquivos = os.listdir(pasta_downloads) # obtem todos os itens da pasta
arquivos = [os.path.join(pasta_downloads, arquivo) for arquivo in arquivos] #gera lista com o caminho completo de todos os itens da pasta
arquivos = sorted(arquivos, key=os.path.getmtime, reverse=True) # organiza a lista de acordo com a data de modificação
ultimo_arquivo = arquivos[0] # obtem o ultimo arquivo baixado
print(ultimo_arquivo)
```

Carregando a base de dados e calculando valores

```python
import pandas as pd
tabela = pd.read_csv(ultimo_arquivo, sep=";")
display(tabela)
# com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos
total_gasto = tabela["ValorUnitário"].sum()
quantidade_produtos = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade_produtos
```

Abrindo gmail e enviando email

```python
import pyperclip

# abra nova guia e abre gmail
# pyautogui.hotkey("ctrl", "t")
pyautogui.hotkey("ctrl", "l") # para digitar o link na mesma aba
pyautogui.write("gmail.com") # digital link do gmail
pyautogui.press("enter") # enter para acessar email
time.sleep(4)
pyautogui.click(x=-1278, y=169) # clica no botão de escrever email
time.sleep(4)
pyautogui.write("radeji7456@jwsuns.com") # digita o destinatario
pyautogui.press("enter") # confirma destinario
pyautogui.press("tab") # muda para o campo de assunto
pyperclip.copy("Relatório de compras") # copia para evitar a falta do caracteres especiais 
pyautogui.hotkey("ctrl", "v") # cola o conteúdo no assunto
mensagem = f"""Prezados,
Segue o relatório de compras

Total gasto: R${total_gasto:,.2f}
Quantidade de produtos: R${quantidade_produtos:,}
Preço médio: R${preco_medio:,.2f}

Qualquer dúvida estou a disposição.

"""
pyautogui.press("tab") # muda para o campo de mensagem
pyperclip.copy(mensagem) # copia mensagem
pyautogui.hotkey("ctrl", "v") # cola mensagem no campo
pyautogui.hotkey("ctrl", "enter") # atalho para enviar mensagem





```
